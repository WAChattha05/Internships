```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
import google.colab as colab
colab.drive.mount('/content/drive')
```

```python
csv_path = '/content/SampleSuperstore.csv'
```

```python
df = pd.read_csv(csv_path)
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for duplicate records
df.duplicated().sum()
```

```python
# Remove any duplicate records
df.drop_duplicates(inplace=True)
```

```python
# Analyze sales, profit, and discount by segment
sns.barplot(x='Segment', y='Sales', data=df)
sns.barplot(x='Segment', y='Profit', data=df)
sns.barplot(x='Segment', y='Discount', data=df)
```

```python

# Analyze sales, profit, and discount by sub-category
sns.barplot(x='Sub-Category', y='Sales', data=df)
sns.barplot(x='Sub-Category', y='Profit', data=df)
sns.barplot(x='Sub-Category', y='Discount', data=df)
```

```python
# Analyze sales, profit, and discount by region
sns.barplot(x='Region', y='Sales', data=df)
sns.barplot(x='Region', y='Profit', data=df)
sns.barplot(x='Region', y='Discount', data=df)
```

```python

# Analyze the relationship between discount and profit
sns.scatterplot(x='Discount', y='Profit', data=df)
```

```python
# Analyze the relationship between discount and profit
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Profit vs Discount')
plt.show()
```
